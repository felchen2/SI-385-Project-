Question: What is the age distribution across different medical conditions?

## Data Link
- https://www.kaggle.com/datasets/abdallaahmed77/healthcare-risk-factors-dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load & Describe Dataset

```python
healthcare_data = pd.read_csv('..\data\dirty_v3_path.csv')
```

```python
healthcare_data.head(15)
print(healthcare_data.shape)
healthcare_data.describe(include = 'all')
```

```python
mean_age = healthcare_data.groupby('Medical Condition')['Age'].mean().reset_index()
print("mean age by medical condition:")
mean_age
```

```python
median_age = healthcare_data.groupby('Medical Condition')['Age'].median().reset_index()
print("median age by medical condition:")
median_age
```

```python
overall_std = healthcare_data.groupby('Medical Condition')['Age'].std(numeric_only=True)
print("Standard Deviation  per Medical Condition:")
print(overall_std)
```

```python
q1 = healthcare_data.groupby('Medical Condition')['Age'].quantile(0.25).reset_index()
print(type(q1))
print(q1)
```

```python
q3 = healthcare_data.groupby('Medical Condition')['Age'].quantile(0.75).reset_index()
print(q3)
```

Written Explanation: 

Some chronic age-related conditions affect older groups, such as hypertension, which has one of the highest median ages, and arthritis. These results align with typical conditions as expected in older age groups. As for conditions that affected younger age groups, asthma presented the youngest median age, which is typical given its prevalence in the younger populations. As for obesity, diabetes, and cancer show medians in the 40s to mid-50s range. This also showed that healthy groups tend to be seen in younger populations, but this group spanned out more to a wider range of age.

## Exploratory Data Analysis

```python
plt.figure(figsize=(10,6))
healthcare_data.boxplot(column="Age", by="Medical Condition")

plt.title("Age Distribution by Medical Condition")
plt.suptitle("")  
plt.xlabel("Medical Condition")
plt.ylabel("Age")
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
